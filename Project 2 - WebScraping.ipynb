## Generate URL list

```python
"""
Generate_URL list
Based on yelp url pattern

"""
with open('cities_scrap.txt','r') as file:
    
    URL_links = []
    for line in file:
        line = line.strip().split(',')
        city = line[0]
        state = line[1]
        
        
        for x in range (35):
            
            if x ==0:
            
                URL = 'https://www.yelp.com/search?find_desc=Restaurants&find_loc='

                # for city more than one word , ex: San Diego
                if ' ' in city:
                    city = city.split(' ')
                    for count, char in enumerate(city):

                        if count ==len(city):
                            URL += char
                        else:
                            URL += char +'%20'
                            
                            
                else:
                    URL += str(city)

                URL +='%2C%20'+ state + '&start='
                golden_url = URL
#                 print(golden_url)
                URL += str(x)
                
            else:
                
                URL = golden_url + str(x*30)  
            
            URL_links.append(URL)

# links:  add 30
# https://www.yelp.com/search?find_desc=Restaurants&find_loc=Willow%20Glen%2C%20San%20Jose%2C%20CA&start=0
# https://www.yelp.com/search?find_desc=Restaurants&find_loc=San%20Francisco%2C%20CA&start=0
# https://www.yelp.com/search?find_desc=Restaurants&find_loc=Palo%20Alto%2C%20CA&start=0
```
